### Задание 3 a  С помощью группировок получите новый датасет data_annual, где будут хранится значения средней температуры за весь период наблюдения для каждого города

### Задание 7 Получите датасет, в котором для каждого индийского города будут содержаться значения года и среднегодовой температуры за этот год. Используя функцию pivot постройте график на котором будут содержаться изменения среднегодовой температуры для всех городов Индии по годам

```python
ind=data[(data['Country']=='India') & (data['Year']>1880)]
ind=ind.groupby(by=['City', 'Year'], as_index=False)['AverageTemperature'].agg(np.mean)
ind=ind.pivot(index='Year', columns='City', values='AverageTemperature')
```

```python
ind
```

```python
ind.plot(figsize=[23,15])
plt.title('I wanna be a data scientist', fontsize=31)
plt.ylim(23,30)
plt.savefig('grafiki.png')
```
